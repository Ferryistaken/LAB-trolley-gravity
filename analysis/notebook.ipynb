```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
one_book_1 = pd.read_csv("/home/ferry/Documents/scuola/3liceo/fisica/lab/carrello-vernier-gravita/measurements/1book/trial1.csv")
one_book_2 = pd.read_csv("/home/ferry/Documents/scuola/3liceo/fisica/lab/carrello-vernier-gravita/measurements/1book/trial2.csv")
one_book_3 = pd.read_csv("/home/ferry/Documents/scuola/3liceo/fisica/lab/carrello-vernier-gravita/measurements/1book/trial3.csv")
print("Data from 1 book dataset loaded")
```

```python
one_book_concat = pd.concat((one_book_1, one_book_2, one_book_3))
one_book_total = one_book_concat.groupby(one_book_concat.index)
one_book_mean = one_book_total.mean()

one_book_mean.head()
```

```python
one_book_mean.plot(kind='line', x='Latest: Position (m)', y='Latest: Velocity (m/s)', color='red', label='line 1')
one_book_mean.plot(kind='line', x='Latest: Velocity (m/s)', y='Latest: Acceleration (m/s²)', color='orange', label='line 2')
```
